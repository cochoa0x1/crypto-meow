```python
cattributes = pd.read_csv('merged.csv')
```

```python
cattributes.sort_values('id').head()
```

```python
cattributes.cattributes = cattributes.cattributes.apply(lambda x: str(x).strip('[]').replace("'",'').split(', '))
```

```python
#count the cattributes
atts={}
for row in cattributes.cattributes:
    for c in row:
        if c not in atts:
            atts[c]=0
        atts[c]+=1
```
